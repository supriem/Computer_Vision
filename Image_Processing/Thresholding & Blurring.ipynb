# Thresholding
is the simplest method of image segmentation. From a grayscale image, thresholding can be used to create binary images  

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('flag.jpeg',0) #0 means is read in grayscale
```

```python
plt.imshow(img)
```

```python
plt.imshow(img,cmap='gray') #in grayscale
```

```python
img.max()
```

```python
cv2.threshold(img,125,255,cv2.THRESH_BINARY) #any value below 125 shifted to 0 and rest to 255
```

```python
ret,thresh1=cv2.threshold(img,125,255,cv2.THRESH_BINARY) #any value below 125 shifted to 0 and rest to 255
```

```python
ret
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
ret,thresh1=cv2.threshold(img,125,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh1,cmap='gray')
```

```python
ret,thresh1=cv2.threshold(img,125,255,cv2.THRESH_TRUNC)
plt.imshow(thresh1,cmap='gray')
```

# Gamma Correction

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_flag():
    img=cv2.imread('flag.jpeg')
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
    
```

```python
load_flag()
```

```python
def display_flag(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_flag()
display_flag(i)
```

```python
gamma = 1/3
```

```python
gamma_corrected=np.power(img,gamma)
plt.imshow(gamma_corrected)
```

```python
display_flag(gamma_corrected)
```

```python
gamma = 1.000001

corrected=np.power(img,gamma)
display_flag(corrected)
```

# Blurring
An image kernel is a small matrix used to apply effects like the ones you might find in Photoshop or Gimp, such as blurring, sharpening, outlining or embossing.

They're also used in machine learning for 'feature extraction', a technique for determining the most important portions of an image

# Gaussian Blur

```python
img=load_flag()
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text='Nepal\'s Flag',org=(15,600),fontFace=font,fontScale=5,color=(200,10,0),thickness=2)

display_flag(img)
```

```python
g_blurred =cv2.GaussianBlur(img,(5,5),10)
```

```python
display_flag(g_blurred)
```

# Median Blur

```python
img=load_flag()
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text='Nepal\'s Flag',org=(15,600),fontFace=font,fontScale=5,color=(200,10,0),thickness=2)
```

```python
median_blurr=cv2.medianBlur(img,5)
```

```python
plt.imshow(median_blurr)
display_flag(median_blurr)
```
