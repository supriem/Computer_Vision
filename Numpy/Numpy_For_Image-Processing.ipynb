Above plot shows 775 px len,1200 width and 3 color channels. 

Every RGB channel alone is just a grayscale.

Each channels has same HW of 3 arrays.

```python
#copying an image
img_c=img_arr.copy()
plt.imshow(img_c) #closer to white means more red
```

cmap=color map

```python
#1=green,0=red,2=blue
#here we have set green color channel so there is no color contribution from green.
img_c[:,:,1]=0
```

```python
plt.imshow(img_c)
```
